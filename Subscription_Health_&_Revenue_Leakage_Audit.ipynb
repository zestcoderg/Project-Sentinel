```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head(4)
```

```python
#Checking for null values in df
df.info()
```

```python
#Empty strings in TotalCharges and MonthlyCharges
print(f"Empty strings in TotalCharges: {len(df[df['TotalCharges'] == ' '])}")
print(f"Empty strings in MonthlyCharges: {len(df[df['MonthlyCharges'] == ' '])}")
```

```python
print(df['Churn'].value_counts(normalize=True))
```

```python
df['TotalCharges']=pd.to_numeric(df['TotalCharges'], errors='coerce')
#df.info()
# df['TotalCharges']=df[df['TotalCharges'].isnull()] # This line caused the error and is removed.
```

```python
print(f"Number of rows where 'TotalCharges' is 0 (originally null): {len(df[df['TotalCharges'] == 0])}")
display(df[df['TotalCharges'] == 0])
```

```python
null_rows = df[df.isnull().any(axis=1)]
print(f"Number of rows with null values: {len(null_rows)}")
display(null_rows)
```

```python
# 1. Locate the specific '0 tenure' rows
# We use the original dataframe or look at where tenure is 0
anomalies = df[df['tenure'] == 0]

# 2. Investigate their status
print(f"Number of customers with 0 tenure: {len(anomalies)}")
print("Are they all 'Non-Churned'? \n", anomalies['Churn'].value_counts())

# 3. Decision Logic:
# Since they are new, 'TotalCharges' being 0 is technically correct for 'historical' spend,
# but it might skew our LTV (Lifetime Value) models.
```

```python
df['TotalCharges']=df['TotalCharges'].fillna(0)
```

```python
df['Churn_encode']=df['Churn'].apply(lambda x:1 if x=='Yes' else 0)

df.head()
```

```python
print("Cleaning Summary:")
print(df[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())
```

```python
leakage_by_contract = df.groupby('Contract').agg({
    'Churn_encode': 'mean',
    'MonthlyCharges': 'sum'
}).rename(columns={'Churn_encode': 'Churn_Rate'})

print(leakage_by_contract)
```

```python
# Grouping by both Contract and InternetService to find the "Risk Zone"
service_risk = df.groupby(['Contract', 'InternetService']).agg({
    'Churn_encode': 'mean',
    'MonthlyCharges': 'count'
}).rename(columns={'Churn_encode': 'Churn_Rate', 'MonthlyCharges': 'Customer_Count'})

print(service_risk.sort_values(by='Churn_Rate', ascending=False))
```

```python
df.head()
```

```python
# Creating a Risk Score based on our findings
def define_risk(row):
    if row['Contract'] == 'Month-to-month' and row['InternetService'] == 'Fiber optic':
        return 'Critical Risk'
    elif row['Contract'] == 'Month-to-month':
        return 'High Risk'
    else:
        return 'Stable'

df['Revenue_Risk_Status'] = df.apply(define_risk, axis=1)

# Saving the file for further analysis
df.to_csv('Project_Sentinel_Cleaned.csv', index=False)

print(" File Exported.")
print(df['Revenue_Risk_Status'].value_counts())
```
